```python
dataset["Administration"].fillna(dataset["Administration"].mean(),inplace=True)
dataset["Marketing Spend"].fillna(dataset["Marketing Spend"].mean(),inplace=True)
```

```python
x = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4:5].values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
x
```

```python
ct = ColumnTransformer([("one",OneHotEncoder(),[3])],remainder ="passthrough")
x = ct.fit_transform(x)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
```

```python
regression = Sequential()
```

```python
regression.add(Dense(units=6, kernel_initializer = "random_uniform", activation="relu"))
```

```python
regression.add(Dense(units=12, kernel_initializer = "random_uniform", activation="relu"))
```

```python
regression.add(Dense(units=12, kernel_initializer = "random_uniform", activation="relu"))
```

```python
regression.add(Dense(units=12, kernel_initializer = "random_uniform", activation="relu"))
```

```python
regression.add(Dense(units=12, kernel_initializer = "random_uniform", activation="relu"))
```

```python
regression.add(Dense(units=1, kernel_initializer = "random_uniform", activation="relu"))
```

```python
regression.compile(optimizer="rmsprop", loss="mse" ,metrics=["accuracy"])
```

```python
regression.fit(x_train, y_train, epochs = 500) 
```

```python
ypred = regression.predict(x_test)
```
